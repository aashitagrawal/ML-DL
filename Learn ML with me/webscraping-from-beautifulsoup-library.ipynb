```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
rank = []
book = []
author = []
rating = []
score = []
for i in range(1,98):
    webpage=requests.get('https://www.goodreads.com/list/show/7.Best_Books_of_the_21st_Century?page={}'.format(i), headers=headers).text
    soup = BeautifulSoup(webpage,'lxml')
    for j in range(0,100):
        rank.append(soup.find_all("td",class_="number")[j].text.strip())
        book.append(soup.find_all("a",class_="bookTitle")[j].text.strip())
        author.append(soup.find_all("a",class_="authorName")[j].text.strip())
        rating.append(soup.find_all("span",class_="greyText smallText uitext")[j].text.strip())
        score.append(soup.find_all("span",class_="smallText uitext")[j].text.strip())
```

```python
dict = {'rank': rank,'book':book,'author':author,'rating':rating,'score':score}
df=pd.DataFrame(dict)
df
```
