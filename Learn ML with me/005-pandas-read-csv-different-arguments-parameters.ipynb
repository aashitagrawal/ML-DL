# 1. Opening a csv file from an URL
Enter the data url in the code snippet below

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

# 11. Handling Dates

```python
pd.read_csv("/kaggle/input/ipldata/IPL.csv", parse_dates=['date']).info()
```

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    if name == "Kings XI Punjab":
        return "KXIP"
    if name == "Delhi Daredevils":
        return "DD"
    if name == "Mumbai Indians":
        return "MI"
    if name == "Kolkata Knight Riders":
        return "KKR"
    else:
        return name
```

# 12. Convertors
to convert long name to short

```python
pd.read_csv("/kaggle/input/ipldata/IPL.csv",converters={'team1':rename, 'team2':rename, 'winner':rename, 'toss_winner':rename})
```
